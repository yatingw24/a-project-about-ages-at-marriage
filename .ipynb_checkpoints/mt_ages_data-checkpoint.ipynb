```python
import matplotlib.pyplot as plt


female_age = 25
male_age = 30


fig, ax = plt.subplots()


ax.bar(['Female', 'Male'], [female_age, male_age], color=['pink', 'blue'])
ax.set_xlabel('Gender')
ax.set_ylabel('Age at First Marriage')
ax.set_title('Age at First Marriage by Gender')


fig.savefig('plot.png')

plt.close(fig)

print("Plot saved as 'plot.png'")
```
